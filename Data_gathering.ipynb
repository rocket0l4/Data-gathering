```python
pd.read_csv('File Name',skiprows = [example => 0,5]) # for skip rows
pd.read_csv('File Name',nrows = 100) # for nrows
```

#Convertors for csv file.

- if we need to change some vlaue in short form(in column), we can use the convertors.

- We can create a function before uploading the dataset.

- In that function we need to specifiy what we want to function do.

- Then we use that function in the convertors.

```python
pd.read_csv('File Name', convertors = {'Column Name ' : function name})
```

```python
dal = BeautifulSoup(webpage,'lxml')
```

```python
imdb = dal.find_all('li',class_ = 'ipc-metadata-list-summary-item sc-10233bc-0 TwzGn cli-parent')
```

```python
len(imdb)
```

```python
names = []
import re
def remove(name2):
    pattern = '[0-9]'
    # dots = '.'
    name2 = [re.sub(pattern, '', i) for i in name2]
    # name = [re.sub(dots,'',i)for i in name]
    return name2

# Driver codeprint(remove(name2))
print(remove(name2))
names.append(remove(name2))
names
```
